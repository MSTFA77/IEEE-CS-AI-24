There's another way to create plots with Matplotlib, it's known as the object-orientated (OO) method. Let's try it.

```python
# Create a plot using plt.subplots()
import matplotlib.pyplot as plt

# Create a figure and an axes using plt.subplots()
fig, ax = plt.subplots()

# Data for the plot
X = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Plotting using the OO method
ax.plot(X, y)

# Add a title and labels using the axes object
ax.set_title('Object-Oriented Plot of X and y')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Display the plot
plt.show()
```

```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)
import numpy as np
import matplotlib.pyplot as plt

# Create an array of 100 evenly spaced numbers between 0 and 100
X = np.linspace(0, 100, 100)

# Compute X squared
Y = X ** 2

# Create a plot
fig, ax = plt.subplots()

# Plot X versus X^2
ax.plot(X, Y)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('Plot of X versus X^2')

# Show the plot
plt.show()
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array of 1000 normally distributed numbers
X = np.random.normal(loc=0, scale=1, size=1000)

# Create a NumPy array of 10 random numbers
X_scatter = np.random.rand(10)
Y_scatter = np.random.rand(10)

# Create a dictionary of favorite foods with their prices
favorite_foods = {
    'Pizza': 12.99,
    'Burger': 8.49,
    'Sushi': 15.75
}

# Extract keys and values from the dictionary
foods = list(favorite_foods.keys())
prices = list(favorite_foods.values())

# Create a figure with 2x2 subplots and figsize of (10, 5)
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Plot X versus X/2 on the top left subplot
axs[0, 0].plot(X, X / 2)
axs[0, 0].set_title('Plot of X versus X/2')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('X/2')

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
axs[0, 1].scatter(X_scatter, Y_scatter)
axs[0, 1].set_title('Scatter Plot of 10 Random Numbers')
axs[0, 1].set_xlabel('X values')
axs[0, 1].set_ylabel('Y values')

# Plot a bar graph of the favorite food keys and values on the bottom left subplot
axs[1, 0].bar(foods, prices, color='skyblue', edgecolor='black')
axs[1, 0].set_title('Bar Graph of Favorite Foods')
axs[1, 0].set_xlabel('Food')
axs[1, 0].set_ylabel('Price ($)')

# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
axs[1, 1].hist(X, bins=30, edgecolor='black')
axs[1, 1].set_title('Histogram of Normally Distributed Data')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Import the CSV file into a DataFrame
car_sales = pd.read_csv('Downloads/car-sales.csv')

# Clean the 'Price' column (removing symbols and converting to numbers)
car_sales['Price'] = car_sales['Price'].replace({'\$': '', ',': ''}, regex=True)
car_sales['Price'] = car_sales['Price'].str[:-2].astype(float)

# Add 'Total Sales' column as a cumulative sum of the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add 'Sale Date' column with successive dates starting from today
today = datetime.today()
car_sales['Sale Date'] = [today + timedelta(days=i) for i in range(len(car_sales))]

# View the updated car_sales DataFrame
print(car_sales.head())
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame if car_sales is not defined
data = {
    'Sale Date': pd.date_range(start='2024-01-01', periods=10, freq='M'),
    'Total Sales': [200, 220, 250, 275, 300, 320, 310, 290, 310, 330]
}
car_sales = pd.DataFrame(data)

# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales', kind='line')

# Show the plot
plt.title('Total Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the following plot:

<img src="../images/matplotlib-heart-disease-chol-age-plot.png"/>

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Define the file path
file_path = 'Downloads/heart-disease.csv'

# Check if the file exists
if os.path.exists(file_path):
    # Import the heart-disease.csv file into a DataFrame
    heart_disease = pd.read_csv(file_path)

    # Step 1: Create DataFrame with patients over 50 years old
    older_patients = heart_disease[heart_disease['age'] > 50]

    # Step 2: Create the plot
    fig, ax = plt.subplots(figsize=(12, 8))

    # Step 3: Plot the data
    older_patients['age'].plot(kind='hist', bins=30, ax=ax, color='skyblue', edgecolor='black', alpha=0.7)

    # Step 4: Customize the plot
    ax.set_title('Histogram of Age for Patients Over 50')
    ax.set_xlabel('Age')
    ax.set_ylabel('Frequency')
    ax.grid(True)

    # Step 5: Add a meanline
    mean_age = older_patients['age'].mean()
    ax.axvline(mean_age, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean Age: {mean_age:.2f}')
    ax.legend()

    # Show the plot
    plt.show()
else:
    print(f"File not found: {file_path}")
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
import pandas as pd
import numpy as np

# Import data if needed
# heart_disease = pd.read_csv('path/to/heart-disease.csv')

# Set the style to 'ggplot' or another available style
plt.style.use('ggplot')

# Example data for demonstration
np.random.seed(0)
age = np.random.randint(50, 80, size=100)
heart_disease = pd.DataFrame({'age': age})

# Create DataFrame with patients over 50 years old
older_patients = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data
older_patients['age'].plot(kind='hist', bins=30, ax=ax, color='skyblue', edgecolor='black', alpha=0.7)

# Customize the plot
ax.set_title('Histogram of Age for Patients Over 50')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
ax.grid(True)

# Add a meanline
mean_age = older_patients['age'].mean()
ax.axvline(mean_age, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean Age: {mean_age:.2f}')
ax.legend()

# Show the plot
plt.show()
```
